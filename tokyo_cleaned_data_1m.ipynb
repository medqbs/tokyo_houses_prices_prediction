```python
# data normalization

def data_norm(df):
    df = (df - df.min())/(df.max()-df.min())
    return df

attributes_list = ["distance_to_station","area","maximum_coverage_ratio","maximum_floor_area_ratio","building_age","price"]

data_norm = data_norm(tokyo_data[attributes_list])

data_norm
```
