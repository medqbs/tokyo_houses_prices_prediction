```python
tokyo_data = pd.read_csv("/content/drive/MyDrive/tokyo_house_price.csv", encoding= 'unicode_escape')

tokyo_data
```

```python
missing_values_count = tokyo_data.isnull().sum()
missing_values_count
```

```python
del_list = ["No","Region","Transaction-price(Unit price m^2)","Purpose of Use",
           "Land shape","Frontage","Layout","Total floor area(m^2)",
           "Frontage roadFDirection","Frontage roadFClassification",
           "Frontage roadFBreadth(m)","City,Town,Ward,Village code",
           "Prefecture","City,Town,Ward,Village","Renovation","Transactional factors"]
```

```python
for i in del_list:
    del tokyo_data[i]
```

```python
tokyo_data
```

```python
tokyo_data["Type"].value_counts()
```

```python
# keep only pre-owned condominiums

tokyo_data.drop(tokyo_data.loc[tokyo_data["Type"] == "Residential Land(Land and Building)"].index, inplace = True)
```

```python
tokyo_data.drop(tokyo_data.loc[tokyo_data["Type"] == "Residential Land(Land Only)"].index, inplace = True)

tokyo_data["Type"].value_counts()
```

```python
del tokyo_data["Use"]
```

```python
tokyo_data.isna().sum()
```

```python
median_coverage_ratio = tokyo_data["Maximus Building Coverage Ratio(%)"].median()
print(median_coverage_ratio)
tokyo_data["Maximus Building Coverage Ratio(%)"] = tokyo_data["Maximus Building Coverage Ratio(%)"].replace(np.nan, median_coverage_ratio)
tokyo_data["Maximus Building Coverage Ratio(%)"] = tokyo_data["Maximus Building Coverage Ratio(%)"]/100
```

```python
median_floor_area_ratio = tokyo_data["Maximus Floor-area Ratio(%)"].median()
print(median_floor_area_ratio)
tokyo_data["Maximus Floor-area Ratio(%)"] = tokyo_data["Maximus Floor-area Ratio(%)"].replace(np.nan, median_floor_area_ratio)
tokyo_data["Maximus Floor-area Ratio(%)"] = tokyo_data["Maximus Floor-area Ratio(%)"]/100
```

```python
tokyo_data.dropna(subset=["City Planning"], axis=0, inplace=True)
```

```python
tokyo_data.drop(tokyo_data.loc[tokyo_data["Nearest stationFDistance(minute)"] == "2H-"].index, inplace = True)
```

```python
tokyo_data["Nearest stationFDistance(minute)"] = tokyo_data["Nearest stationFDistance(minute)"].replace("1H30-2H", 100)
```

```python
tokyo_data["Nearest stationFDistance(minute)"] = tokyo_data["Nearest stationFDistance(minute)"].replace("1H-1H30", 70)
```

```python
tokyo_data["Nearest stationFDistance(minute)"] = tokyo_data["Nearest stationFDistance(minute)"].replace("30-60minutes", 45)
```

```python
median_time = tokyo_data["Nearest stationFDistance(minute)"].median()
print(median_time)
tokyo_data["Nearest stationFDistance(minute)"] = tokyo_data["Nearest stationFDistance(minute)"].replace(np.nan, median_time)
```

```python
tokyo_data.dropna(subset=["Building structure"], axis=0, inplace=True)
```

```python
tokyo_data = tokyo_data.replace("before the war", np.nan)

median_year = tokyo_data["Year of construction"].median()
print(median_year)
tokyo_data["Year of construction"] = tokyo_data["Year of construction"].replace(np.nan, median_year)
```

```python
tokyo_data.isna().sum()
```

```python
tokyo_data
```

```python
tokyo_data.dtypes
```

```python
tokyo_data["Area(m^2)"] = tokyo_data["Area(m^2)"].replace("2,000 m^2 or greater.", 2000)
```

```python
# change the data types

tokyo_data[["Nearest stationFDistance(minute)", "Area(m^2)"]] = tokyo_data[["Nearest stationFDistance(minute)", "Area(m^2)"]].astype("int")
tokyo_data["Year of construction"] = tokyo_data["Year of construction"].astype("int")
tokyo_data["Transaction-price(total)"] = tokyo_data["Transaction-price(total)"].astype("float")

tokyo_data.dtypes
```

```python
# add "Building Age" column

tokyo_data["Building Age"] = 2020 - tokyo_data["Year of construction"]

tokyo_data.head()
```

```python
column_names = list(tokyo_data.columns.values.tolist())
column_names
```

```python
new_column_names = [ 'Type',
                     'Transaction period',
                     'Building structure',
                     'City Planning',
                     'Year of construction',
                     'Area',
                     'Nearest station\x81FName',
                     'Nearest station\x81FDistance(minute)',
                     'Area(m^2)',
                     'Building Age',
                     'Maximus Building Coverage Ratio(%)',
                     'Maximus Floor-area Ratio(%)',
                     'Transaction-price(total)']
tokyo_data = tokyo_data.reindex(columns=new_column_names)
tokyo_data.head()
```

```python
# rename columns

tokyo_data = tokyo_data.rename(columns={"Type":"type","Area":"location",
                                            "Nearest stationFName":"nearest_station",
                                            "Nearest stationFDistance(minute)":"distance_to_station",
                                            "Area(m^2)":"area","Year of construction":"construction_year",
                                            "Building structure":"building_structure",
                                            "City Planning":"city_planning",
                                            "Maximus Building Coverage Ratio(%)":"maximum_coverage_ratio",
                                            "Maximus Floor-area Ratio(%)":"maximum_floor_area_ratio",
                                            "Transaction period":"transaction_period",
                                            "Building Age":"building_age","Transaction-price(total)":"price"})
tokyo_data.head()
```
